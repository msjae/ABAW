# Set up

```python
import os
from glob import glob
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
data_root = "/media/minseongjae/HDD/data/AffWild2/Features"
anno_root = "/media/minseongjae/HDD/data/AffWild2/6th ABAW Annotations"
```

# VA

```python
# task: "EXPR_Recognition_Challenge", "VA_Estimation_Challenge", "AU_Detection_Challenge"
task = "VA_Estimation_Challenge"
# split: "Train", "Validation"
splits = ["Train", "Validation"]
for split in splits:
    # Set anno_paths
    anno_dir = os.path.join(anno_root, f"{task}", f"{split}_Set")
    anno_paths = glob(os.path.join(anno_dir, "*.txt"))

    dfs = []
    for anno_path in tqdm(anno_paths):
        # 첫 번째 행을 제외한 나머지 행을 읽어옴
        with open(anno_path, "r") as f:
            lines = f.readlines()[1:]
        # DataFrame을 생성하기 위한 준비
        dir_name = [anno_path.split("/")[-1].split(".")[0] for _ in range(len(lines) - 1)]  # 파일 이름을 나타내는 리스트
        frame_numbers = range(1, len(lines))  # 첫 번째 행을 제외한 나머지 행에 대한 프레임 번호
        feature_path = [os.path.join(dir_name[i], str(frame_numbers[i]).zfill(5)+'.npy') for i in range(len(lines) - 1)]  # 각 행에 대한 feature path
        annotations = [list(map(float, line.strip().split(','))) for line in lines[1:]]
        # feature path에 파일이 존재하는 경우 idx
        idx = [i for i, path in enumerate(feature_path) if os.path.exists(os.path.join(data_root, path))]
        dir_name = [dir_name[i] for i in idx]
        frame_numbers = [frame_numbers[i] for i in idx]
        feature_path = [feature_path[i] for i in idx]
        annotations = [annotations[i] for i in idx]
        # DataFrame 생성
        df = pd.DataFrame({
            "Directory": dir_name,
            "Frame": frame_numbers,
            "Feature_path": feature_path,
            "Annotation": annotations
        })
        dfs.append(df)
    # 모든 DataFrame을 하나로 합침
    df = pd.concat(dfs, ignore_index=True)
    # DataFrame을 저장
    df.to_csv(f"/media/minseongjae/HDD/data/AffWild2/6th ABAW Annotations/{task}_{split}.csv", index=False)
```

```python
tasks = ["EXPR_Recognition_Challenge", "VA_Estimation_Challenge", "AU_Detection_Challenge"]
splits = ["Train", "Validation"]
disregarded_values = {"EXPR_Recognition_Challenge": -1, "VA_Estimation_Challenge": -5, "AU_Detection_Challenge": -1}
for task in tasks:
    for split in splits:
        df = pd.read_csv(f"/media/minseongjae/HDD/data/AffWild2/6th ABAW Annotations/{task}_{split}.csv", low_memory=False)
        # any value in the Annotation column have disregarded_values[task]
        if task == "EXPR_Recognition_Challenge":
            df = df[df["Annotation"] != disregarded_values[task]]
        else:
            df = df[~df["Annotation"].apply(lambda x: any(val == disregarded_values[task] for val in [float(y.replace('[','').replace(']','').strip()) for y in str(x).split(',')]))]
        df.to_csv(f"/media/minseongjae/HDD/data/AffWild2/6th ABAW Annotations/{task}_{split}.csv", index=False)
        print(f"{task}_{split}: {len(df)}")
```

```python
import albumentations as A
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import numpy as np
from albumentations.pytorch import ToTensorV2
```

```python
transform = A.Compose([
            A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.3),
            A.HorizontalFlip(p=0.5),
            A.GaussianBlur(blur_limit=(3, 7), p=0.3),
            A.Sharpen(alpha=(0.2, 0.5), lightness=(0.5, 1.0), p=0.3),
            A.Resize(224, 224),
            A.Perspective(scale=(0.05, 0.1), p=0.5),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], max_pixel_value=255.0, always_apply=True),
            ToTensorV2()
        ])
```

```python
img_path = '/home/minseongjae/ABAW/0_data/imgs/1-30-1280x720/00001.jpg'
```

```python
pil_img = Image.open(img_path)
cv2_img = cv2.imread(img_path, cv2.IMREAD_RGB)
```

```python
cv2_img.shape
```

```python
# transform
pil_transformed = transform(image=np.array(pil_img))['image']
cv2_transformed = transform(image=cv2_img)['image']
```
